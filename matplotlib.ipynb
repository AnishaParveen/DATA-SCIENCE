```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each line plot
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 subplot grid and plot data in each subplot
plt.figure(figsize=(10, 8))  # Optional: Set the figure size

plt.subplot(2, 2, 1)  # First subplot
plt.plot(x, y1, label='Line 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x, y2, label='Line 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x, y3, label='Line 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x, y4, label='Line 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot 4')
plt.legend()

plt.tight_layout()  # Optional: Adjust the spacing between subplots for a cleaner layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for two datasets (box1 and box2)
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]


plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.title('Box Plot of Two Datasets')
#plt.grid(True)
#plt.show()
```
