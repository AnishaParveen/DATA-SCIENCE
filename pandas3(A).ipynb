```python
d1={'k1':[10,2,37],
  'k2':[4,5,6]
  }
data1=pd.DataFrame(d1)
```

```python
data1
```

```python
d2={'k1':[10,22,37],
  'k4':[41,57,26]
  }
data2=pd.DataFrame(d2)
```

```python
data2

```

```python
pd.merge(data1,data2)#one col must be same
#intersection is happening
```

```python
pd.merge(data1,data2,how='left')#left 
```

```python
pd.merge(data1,data2,how='left', on ='k1')#left 
```

```python
pd.merge(data1,data2,how='left',left_on='k2',right_on='k4')#left 
```

```python
pd.merge(data1,data2,how='right')#right 
```

```python
pd.merge(data1,data2,how='outer')#both outer 
```

```python
pd.merge(data1,data2,how='cross')#all possible combination
```
