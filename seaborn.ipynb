# seaborn

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)
```

```python
sns.displot(iris['sepal_length']) #histograme
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips)
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='smoker')#hue does a groupby operation based upon selected column
```

```python
tips['size'].value_counts()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size')
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style='size')
```

```python
tips['time'].value_counts()
```

```python
sns.displot(tips['time'])
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style='size',hue='time')
```

```python
tips['day'].value_counts()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style='size',hue='day')
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip)#scatterplot + histograme
```

```python
sns.jointplot(tips)
```

```python
sns.pairplot(tips) #based on all pair of number in a data set
```

```python
sns.jointplot(iris)
```

```python
sns.pairplot(iris)
```
